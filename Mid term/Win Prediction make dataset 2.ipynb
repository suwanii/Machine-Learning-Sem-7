```python
import pandas as pd
from collections import Counter
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dftemp_1 = pd.read_csv(r"D:\Semester 7\Machine Learning\Assignment\GlobalLandTemperaturesByCountry.csv")
```

```python
# converting date column to datetime
dftemp_1['Date']= pd.to_datetime(dftemp_1['dt'])
# converting date column to str so it can be split
dftemp_1['Date'] = dftemp_1['Date'].astype(str)
dftemp_1['Year'] = dftemp_1['Date'].apply(lambda x: int(str.split(x,"-")[0]))
dftemp_1['Month'] = dftemp_1['Date'].apply(lambda x: int(str.split(x,"-")[1]))
```

```python
#dftemp_1.to_csv(r"D:\Semester 7\Machine Learning\Assignment\GlobalLandTemperaturesByCountry_YM.csv")
```

```python
# taking only the rows which are after FIFA matches started and dropping columns that are not in need
dftemp_FIFA = dftemp_1.loc[dftemp_1.Year>= 1872].reset_index().drop(columns = ['AverageTemperatureUncertainty','index','dt'])
dftemp_FIFA.head()
```

```python
#getting the yearly average of temperature for each country
dftemp_FIFA_temp_avg = dftemp_FIFA.groupby(['Year','Country']).mean().reset_index().drop(columns = ['Month'])
dftemp_FIFA_temp_avg.head()
```

```python
# getting the list of countries in the temperature dataset
tempcountries1 = dftemp_FIFA.Country.unique().tolist()
```

```python
dfFIFAdata_final = pd.read_csv(r"C:\Users\Suwani Gunasekara\OneDrive\Documents\Sem 7 ML\Mid term\dfFIFAdata_final_wcountry.csv")
dfFIFAdata_final.head()
```

```python
# getting the list of countries played in FIFA World Cup
teamcountries = dfFIFAdata_final.team.unique().tolist()
```

```python
# getting a list of countries which are not in the temp. df
nc = []
for cont in teamcountries:
    if cont not in tempcountries1:
        nc.append(cont)
```

```python
# function which outputs the temperature of the host country and team based on the year
def contTemp(country,tcountry,year):
    # country = country where the tournament is hosted
    # tcountry = country of the playing team
    tempdict = {}
    host_temp = dftemp_FIFA_temp_avg.loc[(dftemp_FIFA_temp_avg.Country == country)&(dftemp_FIFA_temp_avg.Year == year),'AverageTemperature'].iloc[0]
    team_temp = dftemp_FIFA_temp_avg.loc[(dftemp_FIFA_temp_avg.Country == tcountry)&(dftemp_FIFA_temp_avg.Year == year),'AverageTemperature'].iloc[0]
    #away_temp = dftemp_FIFA_temp_avg.loc[(dftemp_FIFA_temp_avg.Country == acountry)&(dftemp_FIFA_temp_avg.Year == year),'AverageTemperature'].iloc[0]
    
    tempdict['host_temp'] = host_temp
    tempdict['team_temp'] = team_temp
    #tempdict['away_temp'] = away_temp
    return tempdict
```

```python
wl = []
for index,row in dfFIFAdata_final.iterrows():
    #print(row)
    
    alldict = {}
    country = row['country']
    tcountry = row['team']
    year     = row['Year']
        
    print(tcountry)
    print(country)
    print(year)
    if year >= 2013:
        temp = contTemp(country,tcountry,2013)
    else:
        temp = contTemp(country,tcountry,year)
   
    print(temp)
    alldict['host_temp'] = temp['host_temp']
    alldict['team_temp'] = temp['team_temp']
    alldict['team']      = row['team']
    alldict['Year']      = row['Year']
    
    wl.append(alldict)
    df1 = pd.DataFrame(wl)
```

```python
# setting the index for year and team
df1.set_index(['Year','team'])
```

```python
dfFIFAdata_final_wtemp = dfFIFAdata_final.merge(df1, on = ['Year','team'])
dfFIFAdata_final_wtemp.head()
```

```python
dfFIFAdata_final_wtemp['temp_diff'] = dfFIFAdata_final_wtemp['team_temp'] - dfFIFAdata_final_wtemp['host_temp']
```
