```python
dfFIFAq  = dfFIFAq.loc[(dfFIFAq.Year >= 1992) & (dfFIFAq.Year <= 2018)].reset_index()
dfFIFAq.head()
```

```python
# function which outputs the population of the team based on the year
def getStat(Hcountry,Acountry,year):
    
    statdict = {}
    HT_Hattack = dfFIFAq.loc[(dfFIFAq['team'] == Hcountry)&(dfFIFAq.FIFAyear == year),'H_AttackS'].iloc[0]
    AT_Hattack = dfFIFAq.loc[(dfFIFAq['team'] == Acountry)&(dfFIFAq.FIFAyear == year),'H_AttackS'].iloc[0]
    
    HT_Aattack = dfFIFAq.loc[(dfFIFAq['team'] == Hcountry)&(dfFIFAq.FIFAyear == year),'A_AttackS'].iloc[0]
    AT_Aattack = dfFIFAq.loc[(dfFIFAq['team'] == Acountry)&(dfFIFAq.FIFAyear == year),'A_AttackS'].iloc[0]
    
    HT_Hdefense = dfFIFAq.loc[(dfFIFAq['team'] == Hcountry)&(dfFIFAq.FIFAyear == year),'H_DefenseS'].iloc[0]
    AT_Hdefense = dfFIFAq.loc[(dfFIFAq['team'] == Acountry)&(dfFIFAq.FIFAyear == year),'H_DefenseS'].iloc[0]
    
    HT_Adefense = dfFIFAq.loc[(dfFIFAq['team'] == Hcountry)&(dfFIFAq.FIFAyear == year),'A_DefenseS'].iloc[0]
    AT_Adefense = dfFIFAq.loc[(dfFIFAq['team'] == Acountry)&(dfFIFAq.FIFAyear == year),'A_DefenseS'].iloc[0]
    
    
    statdict['HT_Hattack'] = HT_Hattack
    statdict['AT_Hattack'] = AT_Hattack
    
    statdict['HT_Aattack'] = HT_Aattack
    statdict['AT_Aattack'] = AT_Aattack
    
    statdict['HT_Hdefense'] = HT_Hdefense
    statdict['AT_Hdefense'] = AT_Hdefense
    
    statdict['HT_Adefense'] = HT_Adefense
    statdict['AT_Adefense'] = AT_Adefense
    
    return statdict
```

```python
wl = []
for index,row in data_FIFA.iterrows():
    #print(row)
    
    alldict = {}
    Hcountry = row['home_team']
    Acountry = row['away_team']
    year     = row['Year']
        
  
    print(year)
    print(Hcountry)
    print(Acountry)
    try:
        stat = getStat(Hcountry,Acountry,year)

        alldict['home_team'] = row['home_team']
        alldict['away_team'] = row['away_team']
        alldict['home_goals']= row['home_goals']
        alldict['away_goals']= row['away_goals']
        alldict['Total_Score_home_team']= row['Total_Score_home_team']
        alldict['Total_Score_away_team']= row['Total_Score_away_team']
        alldict['HomeTeamRank_Difference']= row['HomeTeamRank_Difference']
        alldict['AwayTeamRank_Difference']= row['AwayTeamRank_Difference']
        alldict['Total_count_home_team']= row['Total_count_home_team']
        alldict['Total_count_away_team']= row['Total_count_away_team']
        alldict['Country']   = row['Country']
        alldict['Year']      = row['Year']
        alldict['tournament']= row['tournament']
        alldict['host_temp']= row['host_temp']
        alldict['home_temp']= row['home_temp']
        alldict['away_temp']= row['away_temp']
        alldict['homeTemp_diff']= row['homeTemp_diff']
        alldict['awayTemp_diff']= row['awayTemp_diff']
        alldict['home_pop'] = row['home_pop']
        alldict['away_pop'] = row['away_pop']
        alldict['Hteam_gdp'] = row['Hteam_gdp']
        alldict['Ateam_gdp'] = row['Ateam_gdp']

        alldict['HT_Hattack'] = stat['HT_Hattack']
        alldict['AT_Hattack'] = stat['AT_Hattack']
        alldict['HT_Aattack'] = stat['HT_Aattack']
        alldict['AT_Aattack'] = stat['AT_Aattack']
        alldict['HT_Hdefense'] = stat['HT_Hdefense']
        alldict['AT_Hdefense'] = stat['AT_Hdefense']
        alldict['HT_Adefense'] = stat['HT_Adefense']
        alldict['AT_Adefense'] = stat['AT_Adefense']


        wl.append(alldict)
        df1 = pd.DataFrame(wl)
    except IndexError:
        print("error")

```
