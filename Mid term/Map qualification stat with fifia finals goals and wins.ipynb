```python
# function which outputs the population of the team based on the year
def getStat(team,year):
    
    statdict = {}
    WinCount = dfFIFAdata_final.loc[(dfFIFAdata_final['team'] == team)&(dfFIFAdata_final.Year == year),'WinCount'].iloc[0]
    GoalCount = dfFIFAdata_final.loc[(dfFIFAdata_final['team'] == team)&(dfFIFAdata_final.Year == year),'total_goals'].iloc[0]

    statdict['WinCount'] = WinCount
    statdict['GoalCount'] = GoalCount
    
    return statdict
```

```python
wl = []
for index,row in dfFIFAqdata.iterrows():
    #print(row)
    
    alldict = {}
    team    = row['team']
    year     = row['FIFAyear']
        
  
    print(year)
   
    try:
        stat = getStat(team,year)
        
        alldict['Year'] = row['Year']
        alldict['FIFAyear'] = row['FIFAyear']
        
        alldict['team'] = row['team']
        alldict['H_AttackS'] = row['H_AttackS']
        alldict['A_AttackS'] = row['A_AttackS']
        alldict['H_DefenseS'] = row['H_DefenseS']
        alldict['A_DefenseS'] = row['A_DefenseS']
        alldict['away_goals_conceded'] = row['away_goals_conceded']                
        alldict['away_score'] = row['away_score']                
        alldict['away_games'] = row['away_games']                
        alldict['home_score'] = row['home_score']                
        alldict['home_goals_conceded'] = row['home_goals_conceded']                
        alldict['home_games'] = row['home_games']                
        alldict['hs_mean'] = row['hs_mean']                
        alldict['as_mean'] = row['as_mean']                
        alldict['as_conceded_mean'] = row['as_conceded_mean']                
        alldict['hs_conceded_mean'] = row['hs_conceded_mean']    
        
        alldict['FIFAWinCount'] = stat['WinCount']                
        alldict['FIFAGoalCount'] = stat['GoalCount']                
        
        
       

        wl.append(alldict)
        df1 = pd.DataFrame(wl)
    except IndexError:
        print("error")

```

```python
df1['H_DefenseS^2'] = df1['H_DefenseS'] ** 2
df1['A_DefenseS^2'] = df1['A_DefenseS'] ** 2
df1['A_AttackS^2'] = df1['A_AttackS'] ** 2
df1['H_AttackS^2'] = df1['H_AttackS'] ** 2
```

```python
X = df1[['H_AttackS','A_AttackS','H_DefenseS','A_DefenseS','H_DefenseS^2','A_DefenseS^2','A_AttackS^2','H_AttackS^2']]

y = df1['FIFAWinCount']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Fitting the model
FIFA_model = LinearRegression()
FIFA_model.fit(X_train, y_train)

# Returning the R^2 for the model
FIFA_r2 = FIFA_model.score(X_train, y_train)
print('R^2: {0}'.format(FIFA_r2))
```
